```python
# load the iris dataset as an example
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
wine_data=load_iris()
```

```python
wine_df=pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
wine_df.head()
```

```python
wine_df["target"]=wine_data.target
```

```python
wine_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=wine_df[wine_data.feature_names].copy()
Y=wine_df["target"].copy()
scaler= StandardScaler()
scaler.fit(X)
x_scaled=scaler.transform(X.values)
x_scaled[0]
```

```python
from sklearn.model_selection import train_test_split
x_train_scaled,x_test_scaled,y_train,y_test=train_test_split(x_scaled,Y,train_size=0.7,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression_model=LogisticRegression()
```

```python
logistic_regression_model.fit(x_train_scaled,y_train)
```

```python
predict=logistic_regression_model.predict(x_test_scaled)
predict
```

```python
from sklearn.metrics import classification_report
```

```python
report=classification_report(y_test,predict)
print(report)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
svm_model=SVC()
decision_tree_model=DecisionTreeClassifier()
svm_model.fit(x_train_scaled,y_train)
decision_tree_model.fit(x_train_scaled,y_train)


svm_predict=svm_model.predict(x_test_scaled)
decision_tree_predict=decision_tree_model.predict(x_test_scaled)
```

```python
print(svm_predict)
```

```python
print(decision_tree_predict)
```

```python
from sklearn.metrics import classification_report

model_predict = {
    "Logistic Regression": predict,
    "Support Vector Model": svm_predict,
    "Decision Tree": decision_tree_predict
}

for model, pred in model_predict.items():
    report = classification_report(y_test, pred)
    print(f"Model: {model}\n{report}")
```
